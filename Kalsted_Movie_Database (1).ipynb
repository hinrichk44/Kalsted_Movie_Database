# The Best and Worst Movies

The first analysis we will do is find the best and worst ranked movies in the dataset

```python
from IPython.display import HTML
```

```python
#Here we are creating a subset called "best movies" from the main dataset

best_movies = movies[['poster_path', 'title', 'budget_musd', 'revenue_musd', 
                      'vote_count', 'vote_average', 'popularity']].copy()

best_movies
```

```python
#Here we are creating two new columns to add to the best movies dataframe

#profit is simply revenue minus budget
best_movies['profit_musd'] = movies.revenue_musd.sub(movies.budget_musd)

#return on investment is revenue divided by budget
best_movies['return'] = movies.revenue_musd.div(movies.budget_musd)
```

```python
#Here we are renaming the columns for the best movies dataframe
best_movies.columns = ['', 'Title', 'Budget', 'Revenue', 'Votes', 
                       'Average_Rating', 'Popularity', 'Profit', 'ROI']
```

```python
#Here we are resetting the index of best movies to Title
best_movies.set_index('Title', inplace = True)
```

```python
#Here we are sorting the movies by Average Rating from highest to lowest
#As you can see this isn't entirely perfect since many of these movies only have one vote
#It would be better to measure movies that were seen by a greater amount of people

best_movies.sort_values(by = 'Average_Rating', ascending = False)
```

```python
#Here we are sorting the movies by return on investment from most to least

best_movies.sort_values(by = 'ROI', ascending = False)
```

```python
#Here we are filling in the missing values of the Budget and Vote columns with 0

best_movies.Budget.fillna(0, inplace = True)
best_movies.Votes.fillna(0, inplace = True)
```

```python
#Here we are creating a function that allows us to find the best and worst movies in the database
#The number of best or worst movies will be denoted by "n"
#The criteria is decided by "by"
#We are formatting the best or worst movies in ascending order
#We are setting the minimum budget to 0 and minimum amount of votes to 0 as well

def best_worst(n, by, ascending = False, min_bud = 0, min_votes = 0):
    rankings = best_movies.loc[(best_movies.Budget >= min_bud) & (best_movies.Votes >= min_votes),
                              ['', by]].sort_values(by = by, ascending = ascending).head(n).copy()
    return HTML(rankings.to_html(escape=False))
```

# How Successful Are Movie Franchises?

```python
#Here we are creating a separate dataframe for movie franchises

movies['Franchise'] = movies.belongs_to_collection.notna()
```

```python
#Here we are measuring the average revenue of franchises vs. stand-alone movies
#Looks like franchise movies do indeed raise more revenue on average than stand-alone movies. By quite a lot

movies.groupby('Franchise').revenue_musd.mean()
```

```python
#Here we are creating a separate dataframe for return on investment

movies['ROI'] = movies.revenue_musd.div(movies.budget_musd)
```

```python
#Here we are measuring the median return on investment of franchises vs. stand-alone movies
#Looks like franchise movies have a greater return on investment than stand-alone movies

movies.groupby('Franchise').ROI.median()
```

```python
#Here we are measuring the average budget for franchises vs. stand-alone movies
#Looks like franchise movies are equipped with a bigger budget than stand-alone movies on average

movies.groupby('Franchise').budget_musd.mean()
```

```python
#Here we are measuring the average popularity for franchises vs. stand-alone movies
#Looks like franchise movies are more popular than stand-alone movies

movies.groupby('Franchise').popularity.mean()
```

```python
#Here we are measuring the average votes cast for franchises vs stand-alone movies
#Surprisingly stand-alone movies have more votes cast than franchise movies

movies.groupby('Franchise').vote_average.mean()
```

```python
#Here we are simply aggregating all of the metrics we just went over into one nice dataframe

movies.groupby('Franchise').agg({'budget_musd': 'mean', 'revenue_musd': 'mean', 'vote_average': 'mean',
                                'popularity': 'mean', 'ROI': 'median', 'vote_count': 'mean'})
```

```python
#Here we are finding the 20 top franchises in terms of revenue raised
#The Harry Potter movies have earned the most revenue of all movie franchises

franchises.nlargest(20, ('revenue_musd', 'sum'))
```

```python
#Here we are finding the 20 top franchises in terms of average revenue 
#Avatar is the most, but since there is still only one movie made in that franchise it should not count yet

franchises.nlargest(20, ('revenue_musd', 'mean'))
```

# Who Are the Most Successful Actors?

```python
#Here we are setting the dataframe index to "id"
#Every movie has a specific id attached to it

movies.set_index('id', inplace = True)
```

```python
#Here we are seeing what splitting the cast up for each movie looks like

movies.cast.str.split("|")
```

```python
#We are applying that split to a separate dataframe

actors = movies.cast.str.split("|", expand = True)
```

```python
#Here is what it looks like. id 862 and each movie after has separate columns for each actor who appeared in the movie

actors.head()
```

```python
#Now we are setting the index of this new dataframe to id again
#It looks much smoother. The id is one column and the actors are in rows now

actors.stack().reset_index(level=1, drop = True).to_frame()
```

```python
actors = actors.stack().reset_index(level=1, drop = True).to_frame()
```

```python
#Here we are naming the columns "Actor"

actors.columns = ['Actor']
```

```python
#Here we are merging some columns from the original movies dataset into our actors subset

actors = actors.merge(movies[['title', 'revenue_musd', 'vote_average', 'popularity']],
                     how = 'left', left_index = True, right_index = True)
```

```python
#Here is what the finished project looks like

actors.head()
```

```python
#Here we are plotting the top 20 most active actors
#Bess Flowers has appeared in the most movies in this database

plt.figure(figsize = (12,8))
actors.Actor.value_counts().head(20).plot(kind = 'bar', fontsize = 15)
plt.title('Most Active Actors', fontsize = 20)
plt.ylabel('Number of Movies', fontsize = 15)
plt.show()
```

```python
#Here we are aggregating certain values to analyze actor performance at the box office more in-depth

new_actors = actors.groupby('Actor').agg({'revenue_musd': ['sum','mean'], 'vote_average': ['mean'],
                                                      'popularity': ['mean'], 'Actor': ['count']})

new_actors.columns = ['Total_Revenue', 'Mean_Revenue', 'Mean_Rating', 'Mean_Popularity', 'Total_Movies']
```
